#### Data wrangling process

```python
import pandas as pd
import numpy as np
```

```python
pisa_data = pd.read_csv('/Users/glenndinomangowal/Desktop/Git/pisa_data_exploratory/pisa2012.csv', encoding="cp1252", sep=',')
```

```python
df = pisa_data.copy()
```

```python
df_clean = df.filter(['CNT', 'STIDSTD', 'ST04Q01', 'ST11Q01', 'ST11Q02', 'HISCED', 
                      'ST57Q01', 'ST57Q03', 'ST57Q04','ST57Q05', 'ST57Q06', 'ST57Q02',  
                      'ST93Q01','ST93Q03', 'ST93Q06', 'ST93Q07', 
                      'LMINS', 'MMINS', 'SMINS',
                      'PV1MATH','PV2MATH','PV3MATH','PV4MATH','PV5MATH',
                      'PV1READ','PV2READ','PV3READ','PV4READ','PV5READ',
                      'PV1SCIE','PV2SCIE','PV3SCIE','PV4SCIE','PV5SCIE'])
```

```python
df_clean.rename(columns={'CNT' : 'country',
                         'STIDSTD' : 'student_id',
                         'ST04Q01' : 'gender',
                         'ST11Q01' : 'at_home_mother',
                         'ST11Q02' : 'at_home_father',
                         'HISCED'  : 'parents_education',
                         'ST93Q01' : 'p_giveup_easily',
                         'ST93Q03' : 'p_procrastinate',
                         'ST93Q06' : 'p_perfectionist',
                         'ST93Q07' : 'p_exceed_expectations',
                         'ST57Q01' : 'homework_time',
                         'ST57Q02' : 'guided_homework_time',
                         'ST57Q03' : 'personal_tutor_time',
                         'ST57Q04' : 'commercial_company_tutortime',
                         'ST57Q05' : 'with_parent_studytime',
                         'ST57Q06' : 'with_computer_study_time',
                                                                      }, inplace=True)
                         
```

```python
df_clean.dtypes
```

```python
#creating new columns for the average score for each subjects and total score

df_clean['Average_math_score'] = (df_clean['PV1MATH']+df_clean['PV2MATH']+df_clean['PV3MATH']+df_clean['PV4MATH']+df_clean['PV5MATH'])/5
df_clean['Average_reading_score'] = (df_clean['PV1READ']+df_clean['PV2READ']+df_clean['PV3READ']+df_clean['PV4READ']+df_clean['PV5READ'])/5
df_clean['Average_science_score'] = (df_clean['PV1SCIE']+df_clean['PV2SCIE']+df_clean['PV3SCIE']+df_clean['PV4SCIE']+df_clean['PV5SCIE'])/5
```

```python
df_clean['Average_total_score'] = (df_clean['Average_math_score'] + df_clean['Average_reading_score'] + df_clean['Average_science_score'])/3
```

```python
df_clean['Total_learning_time'] = df_clean['LMINS'] + df_clean['MMINS'] + df_clean['SMINS']
df_clean['Out_of_school_totalstudytime'] = df_clean['homework_time'] + df_clean['guided_homework_time'] + df_clean['personal_tutor_time'] + df_clean['commercial_company_tutortime'] + df_clean['with_parent_studytime'] + df_clean['with_computer_study_time']
```

```python
df_clean.info()
```

```python
df_clean.drop(['PV1MATH', 'PV2MATH','PV3MATH','PV4MATH','PV5MATH',
               'PV1READ','PV2READ','PV3READ','PV4READ','PV5READ',
               'PV1SCIE','PV2SCIE','PV3SCIE','PV4SCIE','PV5SCIE',
               'LMINS','MMINS','SMINS'],axis=1, inplace=True)
```

```python
df_clean.replace({'parents_education': {'ISCED 5A, 6' : 'Bachelors or equivalent',
                            'ISCED 5B' : 'Short-cycle tertiary education',
                            'ISCED 3A, ISCED 4' : 'Post-secondary tertiary',
                            'ISCED 3B, C' : 'Upper-secondary education',
                            'ISCED 2' : 'Lower-secondary education',
                            'ISCED 1' : 'Primary education',
                            'None' : 'Pre-primary eduation'}}, inplace=True)
```

```python
df_clean.parents_education.value_counts()
```

```python
parents = []


for i, row in df_clean.iterrows():
    if row['at_home_mother']== "Yes" and row['at_home_father'] == "Yes":
        parents.append("Yes")
    elif row['at_home_mother']== "Yes" or row['at_home_father'] == "Yes":
        parents.append("Yes")
    else:
        parents.append("No")
        
df_clean['parents_at_home'] = parents
```

```python
df_clean.drop(['at_home_father','at_home_mother'], axis=1, inplace=True)
```
